```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.

import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID column 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN'], axis =1) # This time: only drop 'EIN' and keep 'NAME'
application_df.head()
```

```python
# Determine the number of unique values in each column
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for bining.
application_type = application_df['APPLICATION_TYPE'].value_counts()
application_type.head()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = application_type.loc[application_type.values < 500].index

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()                                
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts_only_1 = classification_counts[classification_counts >1]
classification_counts_only_1.head()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = classification_counts.loc[classification_counts.values < 1000].index

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Anlayse the 'NAME' value counts for binning
    # Binning refers to the process of grouping or categorising data ito discrete intervals.
name_count = application_df['NAME'].value_counts()
name_count.head(30)
```

```python
# Choose a cutoff value and create a list of names to be replaces
names_to_replace = list(name_count[name_count < 100].index)

# Replace in dataframe
for name in names_to_replace:
    application_df['NAME'] = application_df['NAME'].replace(
        name, "Other")
    
# Check to make sure binning was successful
application_df['NAME'].value_counts()
```

```python
# Convert categorical data to numeric with 'pd.get_dummies'
application_numeric = pd.get_dummies(application_df)
application_numeric.head()
```

```python
# Split our preprocessed data into our features and target arrays
# Break up into 'X' and 'y' variables
X = application_numeric.drop("IS_SUCCESSFUL", axis = 1)
y = application_numeric["IS_SUCCESSFUL"]

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
hidden_nodes_layer0 = 80
hidden_nodes_layer1 = 30
hidden_nodes_layer2 = 20

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(
    units = hidden_nodes_layer0,
    input_dim = len(X_train.columns),
    activation = "relu"
))

# Second hidden layer
nn.add(tf.keras.layers.Dense(
    units = hidden_nodes_layer1,
    activation = "relu"
))

# Third hidden layer
nn.add(tf.keras.layers.Dense(
    units = hidden_nodes_layer2,
    activation = "tanh" 
    # "tanh" transforms the output to a range between -1 and 1.
))

# Output layer
nn.add(tf.keras.layers.Dense(
    units = 1,
    activation = "sigmoid"
))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(
    loss = 'binary_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)
```

```python
# Train the model
fit_model = nn.fit(
    X_train_scaled,
    y_train,
    epochs = 100
)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
